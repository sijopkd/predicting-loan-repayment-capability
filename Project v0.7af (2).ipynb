## Data Imputation and Feature Engineering
In this section, we will impute the missing values with 0 or the mean/median depending on the data distrbution. We will also create few columns based on domain knowledge to input to the models

From the plots in the section above, we can see that the most common value in the enquiry field is 0 and hence we will impute all the missing values in these columns with 0

```
#Replace the NA values with 0
application_train.AMT_REQ_CREDIT_BUREAU_HOUR.fillna(0, inplace=True)
application_train.AMT_REQ_CREDIT_BUREAU_DAY.fillna(0, inplace=True)
application_train.AMT_REQ_CREDIT_BUREAU_MON.fillna(0, inplace=True)
application_train.AMT_REQ_CREDIT_BUREAU_WEEK.fillna(0, inplace=True)
application_train.AMT_REQ_CREDIT_BUREAU_QRT.fillna(0, inplace=True)
```

To make the results of the models more intuitive, we will convert all the day values to positive and then further convert them to a yearly format

```
#Changing the days value to positive
application_train['DAYS_EMPLOYED']=abs(application_train['DAYS_EMPLOYED'])
application_train['DAYS_BIRTH']=abs(application_train['DAYS_BIRTH'])

#Converting days to years
application_train['YEARS_EMPLOYED']=application_train['DAYS_EMPLOYED']*1.0/365
application_train['YEARS_BIRTH']=application_train['DAYS_BIRTH']*1.0/365
```

To include the family size into the models, we can bucket the family size into 4 categories, namely single, small, medium and big

```
# Bucketing Family size
application_train['CNT_FAM_MEMBERS'].value_counts()
bins = pd.IntervalIndex.from_tuples([(0,1), (1,2), (2,4), (4,22)])
application_train['FAMILY SIZE'] = pd.cut(application_train['CNT_FAM_MEMBERS'],bins)
application_train['FAMILY SIZE']=application_train['FAMILY SIZE'].astype('str')


application_train['FAMILY SIZE']=application_train['FAMILY SIZE'].replace('(0.0, 1.0]','Single')
application_train['FAMILY SIZE']=application_train['FAMILY SIZE'].replace('(1.0, 2.0]','Small')
application_train['FAMILY SIZE']=application_train['FAMILY SIZE'].replace('(2.0, 4.0]','Medium')
application_train['FAMILY SIZE']=application_train['FAMILY SIZE'].replace('(4.0, 22.0]','Big')
```

We will add 4 columns Credit_Income_Percent, Annuity_Income_Percent, Credit_Term and Days_Employed_Percent which would help to validate our hypothesis

```
# Let us modify few columns based on domain knowledge
# Add domain based features
application_train['CREDIT_INCOME_PERCENT'] = application_train['AMT_CREDIT'] / application_train['AMT_INCOME_TOTAL']
application_train['ANNUITY_INCOME_PERCENT'] = application_train['AMT_ANNUITY'] / application_train['AMT_INCOME_TOTAL']
application_train['CREDIT_TERM'] = application_train['AMT_ANNUITY'] / application_train['AMT_CREDIT']
application_train['DAYS_EMPLOYED_PERCENT'] = application_train['DAYS_EMPLOYED'] / application_train['DAYS_BIRTH']
```

The next step before we can go ahead with building the model would be to convert all the categorical variables to dummy variables using one hot encoding

```
# Convert all the categorical variables to dummy variables
application_train_ohe = pd.get_dummies(application_train)
application_train_ohe.shape
```

We will use an imputer to impute all the other missing values in the columns using the median of the columns as most of the data is skewed

```
# imputer for handling missing values

imputer = Imputer(strategy = 'median')

# Need to impute missing values
app_train_imputed = pd.DataFrame(imputer.fit_transform(application_train_ohe))
app_train_imputed.columns = application_train_ohe.columns
app_train_imputed.head()
```

We will now split the data into a test and train sample with a 20-80% split to fit and evaluate the models obtained

```
# Let us split the data into a train & test sample wit 80-20% split
X = app_train_imputed.drop('TARGET', axis = 1)
Y = app_train_imputed.TARGET

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.20, random_state=0)
```

As we are dealing with a dataset that has a huge class imbalance, we will use the SMOTE (Synthetic Minority Oversampling Technique) concept to solve this issue

```
# Using SMOTE to deal with the class imbalance problem in the dataset

sm = SMOTE(random_state=7)
train_resample, y_resample = sm.fit_sample(X_train,y_train)
#train_resample, y_resample = SMOTE().fit_resample(X_train,y_train)

y_resample = pd.Series(y_resample)
train_resample = pd.DataFrame(train_resample)
X_train.shape, y_train.shape, train_resample.shape, y_resample.shape
```

## Model Building

### 1. Random Forest without resampling

```
# Try Random Forest

# Make the random forest classifier
random_forest = RandomForestClassifier(n_estimators = 100, random_state = 50, verbose = 1, n_jobs = -1)

# Train on the training data
random_forest.fit(X_train, y_train)

# Find the feature importance from the random forest model
feat_importances = pd.Series(random_forest.feature_importances_, index=X_train.columns)

display(feat_importances.nlargest(10).plot(kind='barh'))
```

```
y_pred_rf_prob = random_forest.predict_proba(X_test)[:, 1]
y_pred_rf = random_forest.predict(X_test)
```

```
#Print the confusion matrix and F1-score
confusion_matrix_rf = confusion_matrix(y_test, y_pred_rf)
c_m = pd.DataFrame({'True_pred': confusion_matrix_rf[:, 0], 'False_pred': confusion_matrix_rf[:, 1]}, index = ['True_ground','False_ground'])
print ( 'F1-score', f1_score(y_test, y_pred_rf)*100)
print(c_m)
```

```
print(classification_report(y_test, y_pred_rf))
```

```
#Plot the ROC curve for the random forest model

rf_roc_auc = roc_auc_score(y_test, y_pred_rf_prob)
fpr, tpr, thresholds = roc_curve(y_test, y_pred_rf_prob)
plt.figure()
plt.plot(fpr, tpr, label='Random forest (area = %0.2f)' % rf_roc_auc)
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.savefig('RF_ROC')
plt.show()
display()
```

### 2. Random Forest with SMOTE resampling

### 3. Logistic Regression

```
#Fit the scaled data to a logistic regression model
scaler = MinMaxScaler(feature_range = (0, 1))

scaler.fit(X_train)
train_scaled = scaler.transform(X_train)
test_scaled = scaler.transform(X_test)

log_model = LogisticRegression(random_state=0, multi_class='auto',C=0.1).fit(train_scaled, y_train)
pred_log = log_model.predict_proba(test_scaled)[:,1]
y_pred_log_reg = log_model.predict(test_scaled)

```

```
# Plot the ROC curve for Logistic Regression
log_roc_auc = roc_auc_score(y_test, pred_log)
fpr, tpr, thresholds = roc_curve(y_test, pred_log)
plt.figure()
plt.plot(fpr, tpr, label='Logistic (area = %0.2f)' % log_roc_auc)
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.savefig('Log_ROC')
plt.show()
display()
```

```
#Threshold conversion
threshold = (max(pred_log)/2)
pred_log[pred_log > threshold] = 1
pred_log[pred_log <= threshold] = 0
```

```

#Create and print the the confusion matrix and F1-score
confusion_matrix_log_reg = confusion_matrix(y_test, pred_log)
c_m = pd.DataFrame({'True_pred': confusion_matrix_log_reg[:, 0], 'False_pred': confusion_matrix_log_reg[:, 1]}, index = ['True_ground','False_ground'])
print ( 'F1-score', f1_score(y_test, y_pred_log_reg)*100)
print(c_m)
```

```
# Plot the top 10 most important features obtained from the logistic regression model
cdf = pd.DataFrame(np.transpose(log_model.coef_), X_train.columns, columns=['Coefficients'])

cdf['abs_coef'] = np.where( cdf['Coefficients'] < 0, cdf['Coefficients']*-1, cdf['Coefficients']) 
top_20 = cdf.sort_values(by = 'abs_coef', ascending = False).head(20)

cdf_sparkdf = cdf.reindex(cdf.Coefficients.abs().sort_values(ascending = False).index).reset_index().head(10)
display(spark.createDataFrame(cdf_sparkdf))
```
